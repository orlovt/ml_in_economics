#### Assignment 3 (3 points): download 2 dumps (actors-movie and movie-rating) and produce a pandas dataframe of actor-avgrating using join

```python
# Uploading datasets
names = pd.read_csv('name.basics.tsv', sep = '\t')
rtitles = pd.read_csv('title.ratings.tsv', sep = '\t')
```

```python
names = names[['nconst', 'primaryName', 'knownForTitles']]
```

```python
title_principals = pd.read_csv('title.principals.tsv', sep = '\t')
title_principals = title_principals[['tconst', 'nconst', 'category']]
title_principals = title_principals[title_principals['category'] == 'actor']
```

```python
# joining datatest to produse avarage ratings for all of the movies an actor is listed in
res = pd.merge(title_principals,rtitles,on='tconst',how='outer')
res = res.groupby(by = 'nconst').averageRating.mean()
```

```python
# Taking a look at the result and saving it 
actors_ratings = pd.merge(res, names, on='nconst', how='right')
actors_ratings = actors_ratings[['primaryName', 'nconst', 'averageRating']]
print(actors_ratings.head(10))
actors_ratings.to_csv('task3_res.csv')
```
